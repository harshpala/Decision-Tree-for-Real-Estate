High cardinality means a categorical variable has many unique values, like thousands of different product names. Using label encoding for such columns assigns each category a unique number, which can confuse the model. In price prediction, the model might misinterpret these numbers and make inaccurate predictions. To prevent this, we can use frequency encoding, which replaces each category with how often it appears, or target encoding, which uses the average price for each category. These methods help the model understand the data better and make more reliable price predictions.

#### 6. Task 6:  Handling Imbalanced Data:

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
```

```python
target_variable = TrainData["Price_category"]
feature_set = TrainData.drop(["Price_category", "Price"], axis=1)

under_sampler = RandomUnderSampler(random_state=1)
over_sampler = RandomOverSampler(random_state=1)

features_under, target_under = under_sampler.fit_resample(feature_set, target_variable)

features_over, target_over = over_sampler.fit_resample(feature_set, target_variable)

original_distribution = target_variable.value_counts()
undersampled_distribution = target_under.value_counts()
oversampled_distribution = target_over.value_counts()

distribution_summary = pd.DataFrame({
    "Original Count": original_distribution,
    "Under-Sampled Count": undersampled_distribution,
    "Over-Sampled Count": oversampled_distribution
})

TrainData.drop(columns=['Price_category'], inplace=True)

# Step 8: Display the distribution summary
print(distribution_summary)
```

Oversampling is great for keeping the smaller group balanced without needing more data, but it can make the model too specific to the training data and might not work well when classes are very imbalanced.</br>
Undersampling helps make the model simpler and faster by reducing the size of the larger group, but it can remove important information and reduce the models ability to understand the larger group.</br>

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeRegressor

param_dist = {
    'max_depth': [3, 5, 8, 15, None],
    'min_samples_split': [2, 4, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}

n_iter_search = 20  

rs_orig = RandomizedSearchCV(
    estimator=DecisionTreeRegressor(random_state=1),
    param_distributions=param_dist,
    n_iter=n_iter_search,
    scoring='r2',
    cv=5,
    random_state=1,
    n_jobs=-1 
)

rs_orig.fit(X_train, y_train)

rs_scaled = RandomizedSearchCV(
    estimator=DecisionTreeRegressor(random_state=1),
    param_distributions=param_dist,
    n_iter=n_iter_search,
    scoring='r2',
    cv=5,
    random_state=1,
    n_jobs=-1 
)

rs_scaled.fit(X_train_scaled, y_train_scaled)

print("Best parameters (Original):", rs_orig.best_params_)
print("Best R^2 score (Original):", rs_orig.best_score_)
print("---------------------------------")
print("Best parameters (Scaled):", rs_scaled.best_params_)
print("Best R^2 score (Scaled):", rs_scaled.best_score_)
```

#### 3. Task 3:  Pruning Decison Tree:

```python
# Cost-Complexity Pruning for Scaled Data
prune_path_scaled = rs_scaled.best_estimator_.cost_complexity_pruning_path(X_train_scaled, y_train_scaled)
alphas_scaled, impurities_scaled = prune_path_scaled.ccp_alphas, prune_path_scaled.impurities

# Prune trees using list comprehension
pruned_scaled = [
    DecisionTreeRegressor(random_state=1, ccp_alpha=alpha).fit(X_train_scaled, y_train_scaled)
    for alpha in alphas_scaled
]

# Print details for the most pruned scaled tree
print(
    f"Scaled Tree - Nodes: {pruned_scaled[-1].tree_.node_count}, ccp_alpha: {alphas_scaled[-1]}"
)

# Cost-Complexity Pruning for Original Data
prune_path_orig = rs_orig.best_estimator_.cost_complexity_pruning_path(X_train, y_train)
alphas_orig, impurities_orig = prune_path_orig.ccp_alphas, prune_path_orig.impurities

# Prune trees using list comprehension
pruned_orig = [
    DecisionTreeRegressor(random_state=1, ccp_alpha=alpha).fit(X_train, y_train)
    for alpha in alphas_orig
]
```

```python
# Prepare plotting of learning curves
def plot_curve(model, X, y, X_test, y_test, title, metric):
    train_sizes, train_scores, test_scores = learning_curve(
        model, X, y, cv=5, scoring=metric, train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
    )

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_mean, 'o-', color="blue", label="Training Score")
    plt.plot(train_sizes, test_mean, 'o-', color="red", label="Validation Score")

    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color="blue", alpha=0.1)
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color="red", alpha=0.1)

    plt.title(title)
    plt.xlabel("Training Set Size")
    plt.ylabel(f"{metric.upper()} Score")
    plt.legend(loc="best")
    plt.show()

# Example models (These should be replaced with your actual models)
optimal_pruned_tree_scaled = DecisionTreeRegressor(random_state=1)
dt_scaled = DecisionTreeRegressor(random_state=1)
optimal_pruned_tree_original = DecisionTreeRegressor(random_state=1)
dt_original = DecisionTreeRegressor(random_state=1)

# Replace X_train_scaled, y_train_scaled, X_test, y_test with your actual data splits

# Plot learning curves for the pruned model (scaled data)
plot_curve(optimal_pruned_tree_scaled, X_train_scaled, y_train_scaled, X_test, y_test, 
           "Pruned Decision Tree (Scaled Data)", "r2")

# Plot learning curves for the unpruned model (scaled data)
plot_curve(dt_scaled, X_train_scaled, y_train_scaled, X_test, y_test, 
           "Unpruned Decision Tree (Scaled Data)", "r2")

# Plot learning curves for the pruned model (original data)
plot_curve(optimal_pruned_tree_original, X_train, y_train, X_test, y_test, 
           "Pruned Decision Tree (Original Data)", "r2")

# Plot learning curves for the unpruned model (original data)
plot_curve(dt_original, X_train, y_train, X_test, y_test, 
           "Unpruned Decision Tree (Original Data)", "r2")
```

#### 2.Task 2 :  Residual and Error Analysis:

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_residuals(y_pred, residuals, dataset_type="Test", color='blue'):
    plt.figure(figsize=(10, 6))
    plt.scatter(y_pred, residuals, color=color, alpha=0.5)
    plt.axhline(y=0, color='red', linestyle='--')
    plt.title(f"Residual Plot - {dataset_type} Data")
    plt.xlabel("Predicted Price")
    plt.ylabel("Residuals")
    plt.show()

def plot_residual_distribution(residuals, dataset_type="Test", color='blue'):
    plt.figure(figsize=(10, 6))
    sns.histplot(residuals, kde=True, color=color)
    plt.title(f"Distribution of Residuals - {dataset_type} Data")
    plt.xlabel("Residuals")
    plt.ylabel("Frequency")
    plt.show()

# Assume 'dt_scaled' is your trained DecisionTreeRegressor
# Assume 'X_train_scaled', 'X_test', 'y_train_scaled', 'y_test' are properly prepared

# Predictions
y_pred_test = scaledTree.predict(X_test)
y_pred_train = scaledTree.predict(X_train_scaled)

# Calculate residuals
residuals_test = y_test - y_pred_test
residuals_train = y_train_scaled - y_pred_train

# Plot residuals and their distribution for test data
plot_residuals(y_pred_test, residuals_test, "Test", 'blue')
plot_residual_distribution(residuals_test, "Test", 'blue')

# Plot residuals and their distribution for training data
plot_residuals(y_pred_train, residuals_train, "Training", 'green')
plot_residual_distribution(residuals_train, "Training", 'green')
```
